```python
weather_1 = data_weather[data_weather['Station'] == 1]
weather_2 = data_weather[data_weather['Station'] == 2]
```

```python
weather_1.reset_index(inplace=True)
weather_2.reset_index(inplace=True)
```

```python
weather_1.to_csv(('./input/weather1.csv'))
weather_2.to_csv(('./input/weather2.csv'))
```

```python
weather_1.columns
```

```python
date_series = weather_1.Date
```

```python
weather_1.replace('  T', 0.005, inplace=True)
weather_2.replace('  T', 0.005, inplace=True)
```

```python
drop_cols = ['index', 'Unnamed: 0','Station', 'Date', 'CodeSum']
weather_1.drop(labels=drop_cols, axis=1, inplace=True)
weather_2.drop(labels=drop_cols, axis=1, inplace=True)
```

```python
weather_1 = weather_1.astype('float32')
```

```python
weather_2.replace('-', np.nan, inplace=True)
```

```python
weather_2 = weather_2.astype('float32')
```

```python
weather_2.head()
```

```python
Sunrise_series = weather_1.Sunrise
Sunset_series = weather_1.Sunset
```

```python
weather_1.drop(labels=['Sunset','Sunrise'], axis=1, inplace=True)
weather_2.drop(labels=['Sunset','Sunrise'], axis=1, inplace=True)
```

```python
weather_2.isnull().sum()
```

```python
weather_2.Cool.mean()
```

```python
# Handling nulls
weather_2.Tavg = (weather_2.Tmax + weather_2.Tmin) / 2
```

```python
weather_2.Heat.fillna(0.0, inplace=True)
```

```python
weather_2.Cool.fillna(weather_2.Cool.mean(), inplace=True)
```

```python
weather_1.isnull().sum()
```

```python
weather_2.isnull().sum()
```

```python
weather_1.shape
```

```python
weather_2.shape
```

```python
weather_3 = weather_1.add(weather_2)
```

```python
weather_3 = weather_3 / 2.0
```

```python
weather_3['Date'] = date_series
weather_3['Sunset'] = Sunset_series
weather_3['Sunrise'] = Sunrise_series
```

```python
weather_3.isnull().sum()
```

```python
# Handling nulls
weather_3.WetBulb.fillna(weather_3.WetBulb.mean(), inplace=True)
weather_3.PrecipTotal.fillna(weather_3.PrecipTotal.mean(),inplace=True)
weather_3.StnPressure.fillna(weather_3.StnPressure.mean(),inplace=True)
weather_3.SeaLevel.fillna(weather_3.SeaLevel.mean(),inplace=True)
weather_3.AvgSpeed.fillna(weather_3.AvgSpeed.mean(),inplace=True)
```
