## Creating Dummy for traps with spray data 

This is my attempt to create dummy variables that would show which traps the spraying passed. Results showed that the data was a little sparce to do analysis on.

```python
from math import sin, cos, sqrt, atan2, radians


def distance_finder(lat1, lon1, lat2, lon2):
# approximate radius of earth in m
    R = 6378100.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c

    return distance
```

# Combineing Training rows

This is the fucntion I created in order to combine rows in the taining data based on the date, trap, and species so that the the limit of 50 in the trap would not skew the results

```python
def combine_date_trap(pan):
    d = pan.groupby(['Date', 'Trap', 'Species']).sum()
    d = d.reset_index()
    s = pan.drop_duplicates(['Date', 'Trap', 'Species'])
    s = s.drop(["NumMosquitos", 'WnvPresent'], axis =1)
    s = s.reset_index()
    s['NumMosquitos'] = d['NumMosquitos']
    s['WnvPresent'] = d['WnvPresent']
    return s
```

```python
new_train_combine = combine_date_trap(new_train)
```
